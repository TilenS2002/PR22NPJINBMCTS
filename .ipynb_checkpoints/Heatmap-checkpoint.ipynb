```python

nesreca_ = pd.merge(koordinate, osebe[['ZaporednaStevilkaOsebeVPN', 'PoskodbaUdelezenca']], on='ZaporednaStevilkaOsebeVPN', how='left')
nesreca_ = nesreca_.drop(nesreca_[(nesreca_['GeoKoordinataX'] == 0) & (nesreca_['GeoKoordinataY'] == 0)].index)
nesreca = nesreca.dropna(subset=['PoskodbaUdelezenca'])
nesreca_['PoskodbaUdelezenca'] = nesreca_['PoskodbaUdelezenca'].map({
'BREZ POŠKODBE': 0,
'LAŽJA TELESNA POŠKODBA': 1,
'HUDA TELESNA POŠKODBA': 2,
'SMRT': 3,
'BREZ POŠKODBE-UZ': 4
})
colors = ['green', 'yellow', 'orange', 'red', 'green']
fig, ax = plt.subplots(figsize=(10, 7))

for i, posledica in enumerate(sorted(nesreca['PoskodbaUdelezenca'].unique())):
    subset = nesreca_[nesreca_['PoskodbaUdelezenca'] == posledica]
    alpha = 0.8 if posledica == 3 else 0.3
    ax.scatter(subset['GeoKoordinataX'], subset['GeoKoordinataY'], s=10, alpha=alpha, c=colors[i], label=posledica)
ax.set_title('Zemljevid posledic nesreč')
ax.set_xlabel('GeoKoordinataX')
ax.set_ylabel('GeoKoordinataY')
ax.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

cas_osebe = pd.merge(cas, osebe, on='ZaporednaStevilkaOsebeVPN')
faktor = pd.merge(faktor_cest, faktor_oseb, on='ZaporednaStevilkaOsebeVPN')
data = pd.merge(cas_osebe, cesta, on='ZaporednaStevilkaOsebeVPN')
data = pd.merge(data, faktor, on='ZaporednaStevilkaOsebeVPN')
data = pd.merge(data, koordinate, on='ZaporednaStevilkaOsebeVPN')
data = pd.merge(data, kraj, on='ZaporednaStevilkaOsebeVPN')
data = pd.merge(data, nesreca, on='ZaporednaStevilkaOsebeVPN')
data = pd.merge(data, splosno, on='ZaporednaStevilkaOsebeVPN')
#print(data)


label_encoder = LabelEncoder()
for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = label_encoder.fit_transform(data[col].astype(str))

target = 'KlasifikacijaNesrece'

X = data.drop(columns=[target])
y = data[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)

# Select the first row of the data DataFrame
first_row = data.iloc[5]

# Create a new example with all the attributes except the target attribute
# Create a new example with all the attributes except the target attribute
new_example = pd.DataFrame(columns=[col for col in data.columns if col != target])
new_example = new_example.append(data.iloc[0][new_example.columns], ignore_index=True)
print(new_example)


# Transform the new example into numerical values using the label encoder
for col in new_example.columns:
    if new_example[col].dtype == 'object':
        new_example[col] = label_encoder.transform(new_example[col].astype(str))

# Print the new example
print(new_example.columns)



# Make the prediction on the new example
prediction = clf.predict(new_example)
predicted_label = label_encoder.inverse_transform(prediction)
print(predicted_label)

```
